```python
import numpy as np
import gym
import random
import time
from IPython.display import clear_output
```

```python
env = gym.make("FrozenLake-v0")
```

```python
action_space_size = env.action_space.n
state_space_size = env.observation_space.n

q_table = np.zeros((state_space_size, action_space_size))
print(q_table)
```

```python
num_episodes = 100000
max_steps_per_episode = 1000

learning_rate = 0.1
discount_rate = 0.99

exploration_rate = 1
max_exploration_rate = 1
min_exploration_rate = 0.01
exploration_decay_rate = 0.001
```

```python
rewards_all_episods = []

for episode in range(num_episodes):
    state = env.reset()
    done = False
    rewards_current_episode = 0
    
    for step in range(max_steps_per_episode):
        exploration_rate_threshold = random.uniform(0,1)
        if exploration_rate_threshold > exploration_rate:
            # exploitation
            action = np.argmax(q_table[state, :])
        else:
            # exploration
            action = env.action_space.sample()
        
        new_state, reward, done, info = env.step(action)
        
        # updating Q table
        q_table[state, action] = q_table[state, action] * (1 - learning_rate) + learning_rate * (reward + discount_rate * np.max(q_table[new_state, :]))
        
        state = new_state
        rewards_current_episode += reward
        
        if done == True:
            break
        
    # exploration rate decay
    exploration_rate = min_exploration_rate + (max_exploration_rate - min_exploration_rate) * np.exp(-exploration_decay_rate * episode)

    rewards_all_episods.append(rewards_current_episode)
    
# Average reward each 1000 episode
reward_per_thosand_episodes = np.split(np.array(rewards_all_episods), num_episodes/1000)
count = 1000
print('==== Average reward per thounsand episods ====\n')
for r in reward_per_thosand_episodes:
    print(count , ' : ' , str(sum(r/1000)))
    count += 1000
    
print('==== Updated Q table ====\n')
print(q_table)
```

```python
for episode in range(3):
    state = env.reset()
    done = False
    print('========== EPISODE ',episode+1, ' ============')
    time.sleep(1)
    for step in range(max_steps_per_episode):
        clear_output(wait=True)
        env.render()
        time.sleep(0.3)
        
        action = np.argmax(q_table[state,:])
        new_state, reward, done, info = env.step(action)
        
        if done:
            clear_output(wait=True)
            env.render()
            if reward == 1:
                print('YOU\'VE REACHED THE GOAL !!!')
            else:
                print('YOU FELL THROUH A HOLE :(')
            time.sleep(3)
            clear_output(wait=True)
            break
        state = new_state
    env.close()
```
